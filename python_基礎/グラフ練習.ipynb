```python
x = np.array([1.628, 3.363, 5.145, 7.683, 9.855])
y = np.array([1.257, 3.672, 5.841, 7.951, 9.775])

a = np.array([x, np.ones(x.size)])
a = a.T
m, c = np.linalg.lstsq(a, y, rcond=-1)[0]
t = plt.plot(x, y, 'o')
t = plt.plot(x, (m*x+c))
```

```python
from pylab import *
from numpy import *

x = arange(-3, 10, 0.05)
# shape(x) = (400,)
y = 2.5 * exp(-(x)**2/9) + 3.2 * exp(-(x-0.5)**2/4) + random.normal(0.0, 1.0, len(x))
# shape(y) = (400,)
nParam = 2
A = zeros((len(x), nParam), float)
A[:,0] = exp(-(x)**2/9)
A[:,1] = exp(-(x-0.5)**2/4)
# shape(A) = (400, 2)
(p, residuals, rank, s) = linalg.lstsq(A, y)
# p = array([ 1.84768272,  4.06911899])
# residuals = array([ 228.82195325])
# rank = 2
# s = array([ 10.989164  ,   1.70740154])

plot(x, y, '.')
plot(x, p[0]*A[:,0]+p[1]*A[:,1],'x')

show()
```
