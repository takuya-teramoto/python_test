```python
from matplotlib import gridspec
import matplotlib.pyplot as plt
    
figure = plt.figure(figsize=(20, 8))
gs_master = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 1], width_ratios=[1, 2])

gs_1 = gridspec.GridSpecFromSubplotSpec(nrows=2, ncols=1, subplot_spec=gs_master[0:2, 0])
axes_1 = figure.add_subplot(gs_1[:, :])

gs_2_and_3 = gridspec.GridSpecFromSubplotSpec(nrows=2, ncols=1, subplot_spec=gs_master[0:2, 1])
axes_2 = figure.add_subplot(gs_2_and_3[0, :])
axes_3 = figure.add_subplot(gs_2_and_3[1, :])

gs_4 = gridspec.GridSpecFromSubplotSpec(nrows=1, ncols=2, subplot_spec=gs_master[2, :])
axes_4 = figure.add_subplot(gs_4[:, :])
```

```python
def show_pgraph(year, arr1, arr2, arr3, ymin, ymax, ydim=1):
    #人口ピラミッドの表示
    #表示する人口のindexを得る
    idx = int((year - ymin) / ydim)
    #人口ピラミッドと人口グラフのグリッドを生成
    gs = gridspec.GridSpec(2, 2, height_ratios=(3, 2))
    #グラフの配置を決める
    ax = [plt.subplot(gs[0, 0]), plt.subplot(gs[0, 1]), plt.subplot(gs[1, :])]
    #男性の人口ピラミッドを描く
    ax[0].barh(range(0, 101, 5), arr1[idx], height=3)
    ax[0].set(ylim=(0, 100), xlim=(0, 6000))
    ax[0].invert_xaxis()
    ax[0].yaxis.tick_right()
    #女性の人口ピラミッドを描く
    ax[1].barh(range(0, 101, 5), arr2[idx], height=3)
    ax[1].tick_params(labelleft='off')
    ax[1].set(ylim=(0, 100), xlim=(0, 6000))
    #人口グラフを描き西暦に線を引く
    ax[2].plot(range(ymin, ymax+1, ydim), arr3, ls=":")
    ax[2].plot([year, year], [0, 140000])
    
```

```python
import numpy as np
p_male = np.loadtxt('male_1944_2014.csv', delimiter=",", skiprows=1, usecols=range(1,22))
p_female = np.loadtxt('female_1944_2014.csv', delimiter=",", skiprows=1, usecols=range(1,22))
p_total = p_male + p_female
p_yearly = p_total.sum(axis=1)
```

```python
show_pgraph(1950, p_male, p_female, p_yearly, 1944, 2014)
```

```python
#死亡率
lifechart = np.loadtxt('lifechart2014.csv', delimiter=",", usecols=[3])
#死亡率 - 1
rev_lifechart = np.ones(lifechart.size) - lifechart
#人口置き換え水準相当の出生率
rep_level = np.array([0.0041, 0.107, 0.19, 0.0697, 0.017, 0.0021, 0.0001])
#高位推計に対する出生率
high_rate = np.array([0.0036, 0.0514, 0.1593, 0.0927, 0.0187, 0.0023, 0.0001])
```

```python
#男女の人口データを推定するarrayを定義(最新の2年分の人口データを取得)
fp_male = np.array(p_male[-2:])
fp_female = np.array(p_female[-2:])

#人口置き換え水準に戻る期間
recover_in = 10 #5*5=25年

for i in range(20): #20*5=100年
    new_fp_male = fp_male[-1]
    new_fp_female = fp_female[-1]
    #出生率の設定
    if i > recover_in:
        f_rate = rep_level
    else:
        f_rate = high_rate
    #15-49歳の女性人口に出生率をかけて5年間の総新生児数を計算
    new_born = np.sum(new_fp_female[3:10]*f_rate)*5
    new_fp_male = np.hstack(([new_born*100/205], new_fp_male[:-1]))
    new_fp_female = np.hstack(([new_born*105/205], new_fp_female[:-1]))
    #各階級の人口に死亡率を適用
    new_fp_male *= rev_lifechart
    new_fp_female *= rev_lifechart
    #新しい推定人口を追加
    fp_male = np.vstack((fp_male, new_fp_male))
    fp_female = np.vstack((fp_female, new_fp_female))
    
#男女合算の5歳階級別人口、五年毎の推定総人口のarrayを作る
fp_total = np.round(fp_male + fp_female)
fp_sum = np.sum(fp_total, axis=1)

t = plt.plot(range(2013, 2120, 5), fp_sum)
t = plt.ylim([0, 140000])
t = plt.xlim([2000, 2120])
plt.grid(True)
```
